```python
import pandas as pd

# Read CSV file into a DataFrame
stations_df = pd.read_csv('citybikes_stations.csv')

# Now, df contains your data from the CSV file
stations_df
```

```python
stations_df.info()
```

```python
stations_df.describe()
```

```python
print(stations_df.isnull().sum())
```

```python
stations_df.shape
```

```python
print(stations_df.duplicated().sum())
```

```python
# Identify outliers using IQR
Q1 = stations_df['Bikes Available'].quantile(0.25)
Q3 = stations_df['Bikes Available'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_clean = stations_df[(stations_df['Bikes Available'] >= lower_bound) & (stations_df['Bikes Available'] <= upper_bound)]
```

```python
df_clean
```

```python
stations_df['Latitude'] = stations_df['Latitude'].round(6)
stations_df['Longitude'] = stations_df['Longitude'].round(6)
```

```python
stations_df.to_csv('citybikes_stations.csv', index=False) 
```

```python
restaurants_df = pd.read_csv('correct_yelp_restaurants.csv')
```

```python
print(restaurants_df.isnull().sum())
```

```python
columns_to_fill = ['Address', 'Phone']
restaurants_df[columns_to_fill] = restaurants_df[columns_to_fill].fillna('NA')
```

```python
df_clean_venues
```

```python
restaurants_df['Latitude'] = restaurants_df['Latitude'].round(6)
restaurants_df['Longitude'] = restaurants_df['Longitude'].round(6)
```
